## Webscrapping postal codes, boroughs and neighborhood

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
data_pool = []
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
page = requests.get(url).text
soup = BeautifulSoup(page, 'html.parser')
for table in soup.find_all("table"):
    for tbody in table.find_all("tbody"):
        for row in tbody.find_all("tr"):
            for col in row.find_all("td"):
                for b in col.find_all("b"):
                    data_pool.append(b.text)
                    for span in col.find_all("span"):
                        for a in span.find_all("a"):
                            data_pool.append(a.text)
j = 0
i = 0
zip_codes =[]
while j < len(data_pool)-1:
    if len(data_pool[j])==3:
        i = j+1
        k = 0
        while i<len(data_pool):
            if data_pool[j]!= data_pool[i]:
                k = k+1
            i = i+1
    if k == len(data_pool)-j-1:
        if len(data_pool[j+1])>3:
            zip_codes.append(data_pool[j])
    j = j+1
j = 0
borough =[]
neighborough =[]
while j < len(data_pool)-1:
    i = 0
    while i < len(zip_codes):
        k = 1
        c = []
        if data_pool[j] == zip_codes[i]:
            borough.append(data_pool[j+1])
            while len(data_pool[j+1+k])!=3:
                c.append(data_pool[j+1+k])
                k = k+1
            if len(c)==0:
                neighborough.append(data_pool[j+1])
            else:
                neighborough.append(c)
        i = i+1
    j = j+1
column_names = ['PostalCode', 'Borough', 'Neighborhood']
table_panda = pd.DataFrame(columns=column_names)

d = {'PostalCode': zip_codes, 'Borough': borough, 'Neighborhood': neighborough}
df = pd.DataFrame(data=d)
df
```
